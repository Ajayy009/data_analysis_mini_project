```python
import pandas as pd
import matplotlib.pyplot as plt

# 1.  Creation of fake IPL dataset  
data = {
    "season": [2008, 2008, 2009, 2010, 2010, 2011, 2012, 2012, 2013, 2013],
    "winner": [
        "RR", "CSK", "DC", "CSK", "MI", "CSK", "KKR", "MI", "MI", "CSK"
    ]
}

df = pd.DataFrame(data)

# 2. Dashboard Cards (summary info)
total_matches = len(df)
total_seasons = df["season"].nunique()
top_team = df["winner"].value_counts().idxmax()
top_wins = df["winner"].value_counts().max()

print("üèè IPL Mini Dashboard")
print(f"üìå Total Matches: {total_matches}")
print(f"üìå Seasons Covered: {total_seasons}")
print(f"üèÜ Most Successful Team: {top_team} ({top_wins} wins)")

# 3. Bar chart ‚Äì Wins by Team
wins = df["winner"].value_counts()
plt.figure(figsize=(6,4))
wins.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Wins by Team")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.show()

# 4. Line chart ‚Äì Matches per Season
matches_per_season = df.groupby("season").size()
plt.figure(figsize=(6,4))
matches_per_season.plot(kind="line", marker="o", color="green")
plt.title("Matches per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()
```
