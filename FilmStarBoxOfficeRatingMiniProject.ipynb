```python
import matplotlib.pyplot as plt
import numpy as np

# Fake dataset: Mahesh Babu Movies (example data)
movies = ["Okkadu", "Pokiri", "Dookudu", "Srimanthudu", "Bharat Ane Nenu", "Maharshi"]
years = [2003, 2006, 2011, 2015, 2018, 2019]
ratings = [7.8, 8.2, 8.5, 7.6, 7.9, 7.4]
collections = [25, 60, 100, 85, 90, 95]  # crores
budget = [12, 20, 35, 40, 50, 55]

# Create dashboard figure
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Mahesh Babu Filmography Dashboard ", fontsize=16, fontweight="bold")

# 1. Histogram (distribution of ratings)
axs[0,0].hist(ratings, bins=5, color="skyblue", edgecolor="black")
axs[0,0].set_title("Ratings Distribution")
axs[0,0].set_xlabel("Rating")
axs[0,0].set_ylabel("Frequency")

# 2. Line Plot (year vs ratings)
axs[0,1].plot(years, ratings, marker="o", color="green")
axs[0,1].set_title("Ratings Over Time")
axs[0,1].set_xlabel("Year")
axs[0,1].set_ylabel("Rating")
axs[0,1].grid(True)

# 3. Bar Chart (collections)
axs[1,0].bar(movies, collections, color="orange")
axs[1,0].set_title("Box Office Collections (Cr)")
axs[1,0].set_ylabel("Collections")

# 4. Scatter Plot (budget vs collections)
axs[1,1].scatter(budget, collections, c="red", s=100, alpha=0.7)
for i, m in enumerate(movies):  # label each point with movie name
    axs[1,1].text(budget[i]+1, collections[i], m, fontsize=8)
axs[1,1].set_title("Budget vs Collection")
axs[1,1].set_xlabel("Budget (Cr)")
axs[1,1].set_ylabel("Collections (Cr)")

plt.tight_layout()
plt.show()
```
