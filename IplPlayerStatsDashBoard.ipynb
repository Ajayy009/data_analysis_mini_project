```python
import matplotlib.pyplot as plt
```

```python
# Sample IPL dataset
data = {
    "Player": ["Virat Kohli", "MS Dhoni", "Rohit Sharma", "David Warner", "AB de Villiers",
               "Suresh Raina", "Shikhar Dhawan", "KL Rahul", "Chris Gayle", "Hardik Pandya"],
    "Runs": [973, 455, 523, 641, 687, 548, 451, 593, 608, 402],
    "StrikeRate": [152, 135, 139, 144, 151, 136, 127, 138, 156, 145],
    "Fifties": [7, 3, 4, 6, 5, 4, 3, 5, 4, 2],
    "Centuries": [4, 0, 1, 1, 1, 0, 0, 2, 2, 0],
    "Team": ["RCB", "CSK", "MI", "SRH", "RCB", "CSK", "SRH", "PBKS", "RCB", "MI"]
}

df = pd.DataFrame(data)
print(df.head())
# Set the aesthetic style of the plots
plt.figure(figsize=(10,6)) 
sns.barplot(x="Runs", y="Player", data=df, palette="viridis")
plt.title("Top Run Scorers in IPL (Sample Data)", fontsize=14)
plt.show()
plt.figure(figsize=(8,6))

# Scatter plot of Runs vs Strike Rate
sns.scatterplot(x="Runs", y="StrikeRate", hue="Team", size="Centuries", data=df, palette="Set2", sizes=(50,200))
plt.title("Runs vs Strike Rate (colored by Team, bubble size = Centuries)", fontsize=14)
plt.show()

# Box plot of Runs by Team
plt.figure(figsize=(8,6))
sns.boxplot(x="Team", y="Runs", data=df, palette="Set3")
plt.title("Team-wise Runs Distribution", fontsize=14)
plt.show()

# Heatmap of correlations
plt.figure(figsize=(6,4))
sns.heatmap(df[["Runs","StrikeRate","Fifties","Centuries"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```
